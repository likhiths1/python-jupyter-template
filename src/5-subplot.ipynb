### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.\
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
categories = ['A', 'B', 'C', 'D']
values = [5, 3, 9, 6]
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [10, 15, 7, 5]

# Error handling
if len(x) != len(y):
    raise ValueError("Error: 'x' and 'y' must have the same number of elements.")
if len(categories) != len(values):
    raise ValueError("Error: 'categories' and 'values' must have the same number of elements.")
if len(labels) != len(sizes):
    raise ValueError("Error: 'labels' and 'sizes' must have the same number of elements.")

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# First subplot: Line plot
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = x')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Second subplot: Bar plot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Plot')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)
axs[0, 1].tick_params(axis='x', rotation=45)

# Third subplot: Scatter plot
axs[1, 0].scatter(x, y, color='orange', marker='s', label='Data Points')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Fourth subplot: Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, 
              colors=['gold', 'lightcoral', 'lightblue', 'lightgreen'], 
              wedgeprops={'edgecolor': 'black'})
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
