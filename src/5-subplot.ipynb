### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Data
x, y = [0, 1, 2, 3, 4], [0, 1, 4, 9, 16]
categories, values = ['A', 'B', 'C', 'D'], [5, 3, 9, 6]
labels, sizes = ['Apple', 'Banana', 'Cherry', 'Date'], [10, 15, 7, 5]

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
(ax1, ax2, ax3, ax4) = axs.flat

# Line Plot
ax1.plot(x, y, 'bo-', label='y = x²')
ax1.set(title='Line Plot', xlabel='X-axis', ylabel='Y-axis')
ax1.legend()
ax1.grid(True)

# Bar Plot
ax2.bar(categories, values, color=['red', 'blue', 'green', 'purple'])
ax2.set(title='Bar Plot', xlabel='Categories', ylabel='Values')
ax2.grid(axis='y', linestyle='--', alpha=0.7)

# Scatter Plot
ax3.scatter(x, y, color='orange', marker='s', label='Data Points')
ax3.set(title='Scatter Plot', xlabel='X-axis', ylabel='Y-axis')
ax3.legend()
ax3.grid(True)

# Pie Chart
ax4.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140,
        colors=['gold', 'lightcoral', 'lightblue', 'lightgreen'],
        wedgeprops={'edgecolor': 'black'})
ax4.set_title('Pie Chart')

plt.tight_layout()
plt.show()

```
