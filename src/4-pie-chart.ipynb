### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Error handling
if len(labels) != len(sizes):
    raise ValueError("Error: 'labels' and 'sizes' must have the same number of elements.")

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, 
        colors=['skyblue', 'orange', 'green', 'purple'], wedgeprops={'edgecolor': 'black'})

# Corrected title
plt.title('Programming Languages Popularity')

# Added legend for better readability
plt.legend(labels, loc="best")

# Show the plot
plt.show()
```
