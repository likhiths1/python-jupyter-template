### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Error handling
if len(categories) != len(values):
    raise ValueError("Error: 'categories' and 'values' must have the same number of elements.")
if not all(isinstance(i, (int, float)) for i in values):
    raise TypeError("Error: Values must be numeric (int or float).")

# Create a bar plot
plt.figure(figsize=(6, 4))
bars = plt.bar(categories, values, color=['red', 'blue', 'green', 'purple'])

# Add value labels on top of each bar
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), str(bar.get_height()), 
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Labeling the axes and the plot
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.xticks(rotation=45)  # Rotate labels for better readability if needed

# Show the plot
plt.show()
```
