### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Error handling
if len(x) != len(y):
    raise ValueError("Error: x and y must have the same number of elements.")

# Create a scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='blue', marker='o', label='Data Points')

# Add annotations to each point
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", 
                 xytext=(5,5), ha='right', fontsize=10)

# Labeling the axes and the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot with Annotations')
plt.legend()  # Added legend
plt.grid(True)  # Added grid for better visualization

# Show the plot
plt.show()
```
